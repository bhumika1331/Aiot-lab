```python
data = pd.read_csv('cleaned_file_soical1.csv')
Q1 = data['time_spent'].quantile(0.25)
Q3 = data['income'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Analyzing outliers
outliers = data[(data['time_spent'] < lower_bound) | (data['income'] > upper_bound) ]
print(f"Number of potential outliers: {len(outliers)}")
data['time_spent'] = np.clip(data['time_spent'], lower_bound, upper_bound)
```
